## Use automatic verification dataset

```
from keras.models import Sequential
```

```
from keras.layers import Dense
```

```
import numpy 
```

```
# fix random seed for same set of random values every time 
seed = 7
numpy.random.seed(seed)
```

```
# load dataset
import pandas as pd
from google.colab import files 
uploaded = files.upload()

import io
data = io.BytesIO(uploaded["ionosphere.data"])
df = pd.read_csv(data, header=None)
df.head()
```

```
# split the dataset in x and y
x = df.iloc[:,0:33].values 
y = df.iloc[:,-1].values

# one hot endcoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y
```

```
# create model 
model = Sequential()
model.add(Dense(12, input_dim=33, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu'))
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

# compile model 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics= ['accuracy'])

# fit the model 
model.fit(x, y, validation_split=0.33, epochs = 150, batch_size = 10)
```

```
# evaluate the model 
scores = model.evaluate(x,y)
print("%s: %.2f%%" % (model.model.metrics_names[1], scores[1]*100))
```

## Using a Manual verification Dataset

```
from keras.models import Sequential
from keras.layers import Dense 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy 
import pandas as pd
import io

seed = 7
numpy.random.seed(seed)

from google.colab import files
uploaded = files.upload()
data = io.BytesIO(uploaded["ionosphere.data"])
df = pd.read_csv(data, header = None)
df.head()

x = df.iloc[:, 0:33]
y = df.iloc[:, -1]

encoder = LabelEncoder()
y = encoder.fit_transform(y)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=seed)

model = Sequential()
model.add(Dense(12, input_dim=33, kernel_initializer="uniform", activation= 'relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=150, batch_size=10)

```

```
# evaluate the model 
scores = model.evaluate(x,y)
print("%s: %.2f%%" % (model.model.metrics_names[1], scores[1]*100))
```

## Using manual k-fold cross validation  (THE GOLD STANDARD TECHNIQUE)

```
from keras.models import Sequential 
from keras.layers import Dense 
from sklearn.model_selection import StratifiedKFold
import numpy 
import pandas as pd
import io

seed = 7 
numpy.random.seed(seed)

from google.colab import files
uploaded = files.upload()
data = io.BytesIO(uploaded["ionosphere.data"])
df = pd.read_csv(data, header = None)
df.head()

x = df.iloc[:,0:33].values
y = df.iloc[:,-1].values

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)

kfold = StratifiedKFold(n_splits = 10, shuffle=True, random_state=seed)
cvscores=[]
for i, (train, test) in enumerate(kfold.split(x, y)):
  # create model
  model = Sequential()
  model.add(Dense(12, input_dim=33, kernel_initializer='uniform', activation='relu'))
  model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
  model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
  model.fit(x[train], y[train], epochs=150, batch_size=10, verbose=0)
  
  # evaluate the model 
  scores = model.evaluate(x[test], y[test], verbose=0)
  
  print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
  
  cvscores.append(scores[1] * 100)

  
print("%.2f%% (+/- %.2f%%)" % (numpy.mean(cvscores), numpy.std(cvscores)))
  
  
  
```
